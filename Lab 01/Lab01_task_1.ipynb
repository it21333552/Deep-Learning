```python
import matplotlib.pyplot as plt
import numpy as np
exp_data = np.random.exponential(scale = 1.0, size = 10000)
uniform_data = np.random.rand(100000)
normal_data = np.random.rand(100000)
plt.hist(exp_data, bins = 100)
plt.hist(uniform_data, bins = 1000)
plt.hist(normal_data, bins = 100)
plt.show()


plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue",
label="Expotential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="red",
label="Uniform")
plt.hist(normal_data, density=True, bins=1, histtype="step", color="green",
label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate Z values
z = x**2 + y**2

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_title("3D Plot of Z = X^2 + Y^2")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")


ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([0, 50])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Display the first few rows to understand the data
print(df.head())


features = df[['Defense', 'Speed', 'Stage', 'Total', 'HP', 'Attack']]

# Drop rows with missing values to ensure the correlation calculations are valid
features = features.dropna()

# Calculate Pearson correlation
pearson_corr = features.corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = features.corr(method='spearman')

# Plot heatmaps
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Rank Correlation Heatmap')

plt.show()
```
